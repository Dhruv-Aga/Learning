```python
import numpy as np
import cv2
```

```python
# cap = cv.VideoCapture('')
img = cv2.imread('./.temp/typewriter.jpg')
img.shape
```

```python
cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
all_rows = open('./.temp/synset_words.txt').read().strip().split("\n")
classes = [r[r.find(' ') + 1:] for r in all_rows]
for (i,c) in enumerate(classes):
    if(i == 4):
        break
    print(i,c)
```

```python
net = cv2.dnn.readNetFromCaffe('./.temp/bvlc_googlenet.prototxt', './.temp/bvlc_googlenet.caffemodel')
```

```python
blob = cv2.dnn.blobFromImage(img, 1, (224, 224))
net.setInput(blob)
outp = net.forward()
print(max(outp[0]))
```

```python
idx = np.argsort(outp[0])[::-1]
print(idx[:5])
```

```python
for (i, id) in enumerate(idx):
    print('{}. {} ({}): Probability {:.1}%'.format(i+1, classes[id], id, outp[0][id]))
```

```python
cap = cv2.VideoCapture('./.temp/shore.mov')
```

```python
if cap.isOpened() == False:
    print('Cannot open file or video stream')
while True:
    ret, frame = cap.read()
    
    if ret == True:
        cv2.imshow('Frame', frame)
        
        if cv2.waitKey(25) & 0xFF == 27:
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
while True:
    ret, frame = cap.read()
    blob = cv2.dnn.blobFromImage(frame, 1, (224, 224))
    net.setInput(blob)
    outp = net.forward()
    r = 0
    for i in np.argsort(outp[0])[::-1][:5]:
        txt = ' "%s" probability "%.3f" ' % (classes[i], outp[0][i] * 100)
        cv2.putText(frame, txt, (0, 25 + 40*r), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
        r+=1
    if ret == True:
        cv2.imshow('Frame', frame)
        
        if cv2.waitKey(25) & 0xFF == 27:
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```
